Inplace Operations</p>

If you make a change to a dataframe and you want it to be permanent, you must save it back in the dataframe. For example,

bigfoot.drop_duplicates() 

will show you the result of dropping duplicates when you run the command, but it will not save the change in the bigfoot dataframe.

To save the change, you must do:

bigfoot = bigfoot.drop_duplicates()

An alternative that you will often see online is to use the inplace keyword. This keyword means to make the change permanently without requiring you to save the change manually with the equals sign. The example below does the same thing as the prior code cell.

```python
bigfoot.drop_duplicates(inplace=True)
```

Creating New Columns using Conditions</p>

There are many times where you might want to create a new value in a row based on values in another row. For example, maybe given a month you want to output the season, or given a temp_f you want to categorize the day as cold, mild, or hot. The apply function helps us do this.

```python
def get_season(month):
    if month <= 3:
        return 'Winter'
    elif month <= 6:
        return 'Spring'
    elif month <= 9:
        return 'Summer'
    else:
        return 'Fall'

# Provide the new column name on left side below. On the right side, tell Python what column you'll be calling 
# the function on, then use apply to call it.

bigfoot_valid['season'] = bigfoot_valid['month'].apply(get_season)
```

Let's see if it worked by grabbing the first five rows of data. Let's only look at the month and the season that it calculated.

```python
bigfoot_valid[['month','season']].head()
```

Now let's bin the temperatures into hot, mild, and cold categories.

```python
def set_temperature_category(num):
    if num >= 80:
        return 'Hot'
    elif num >= 40:
        return 'Mild'
    else:
        return 'Cold'

# This calls the function. On the left side, we specify the name of the new column.
# On the right side, we specify the column that we're using, then call the apply function and provide 
# it the function name that we're calling.
bigfoot_valid['temp_category'] = bigfoot_valid['temp_f'].apply(set_temperature_category)
```

You can also use the sum function. It doesn't make much sense to use here, but this example calculates the sum of all temp_f values by season.

```python
bigfoot_valid.groupby('season')['temp_f'].sum()
```

We can break results down by several columns if desired.

How many sightings are there for each season and classification? Unstack formats the table in a nicer format.

```python
class_by_season_counts = bigfoot_valid.groupby(['classification','season'])['temp_f'].mean().unstack()
class_by_season_counts
```

```python
class_by_season_info = bigfoot_valid.groupby(['classification','season'])['temp_f'].agg(["max", "min"])
class_by_season_info
```

For numerical values, we can use the cut function to create bins for the values. If we want equally spaced bins, we can specify the number of bins:

```python
bigfoot_valid['temp_bin'] = pd.cut(x=bigfoot_valid['temp_f'], bins=3, right=True).head()
```

Unfortunately this often uses fractional values that don't look nice in an output table. Let's try specifying the bin boundaries instead:

```python
bigfoot_valid['temp_bin'] = pd.cut(x=bigfoot_valid['temp_f'], bins=[-10, 20, 50, 80, 110],
                    labels=['-10 to 20', '20 to 50', '50 to 80', '80 to 110',])
```

There's also a function called qcut that can cut the data based on quantiles (the distribution of the data) rather than just cutting into constant sized bins based on the min and max values.

```python
bigfoot_valid['temp_quantiles'] = pd.qcut(bigfoot_valid['temp_f'], q=5, labels=['0-20%',
                                '20-40%',
                                '40-60%', 
                                '60-80%',
                                '80-100%'])
# Only need to see the two relevant columns to determine if it worked correctly.
bigfoot_valid[['temp_f','temp_quantiles']].head(10)
```

How many times were temperatures in each bin observed by season?

```python
class_by_temp_bin_count = bigfoot_valid.groupby(['temp_bin', 'season'])['temp_bin'].count().unstack()
class_by_temp_bin_count
```

You can combine filters and groupbys. One approach is to create a filter and save it to a new dataframe, then use groupby on the new dataframe. Alternatively, if you want a quick calculation, you could do it all in one line.

Let's do the same calculation as above, but only for NJ observations.

```python
class_by_temp_bin_nj_count = bigfoot_valid[bigfoot_valid['state']=='New Jersey'].groupby(['temp_bin', 'season'])['temp_bin'].count().unstack()
class_by_temp_bin_nj_count
```

<h1 style="font-size:2em;color:#2467C0">Pivot Tables</h1>

Pandas has a similar functionality called pivot tables, which are akin to their usage in Excel. The main difference is how the output is formatted by default. There's a longer tutorial of pivot tables at https://jakevdp.github.io/PythonDataScienceHandbook/03.09-pivot-tables.html. Generally speaking, there's not a strong reason in this class to use one over the other. 

The following two examples yield the identical result.

```python
bigfoot_valid.groupby(['classification','season'])['temp_f'].mean().unstack()
bigfoot_valid.pivot_table('temp_f', index='classification', columns='season')
```

You can have multilevel pivot tables, just as you could for groupbys. Here is an example using groupby followed by the same using a pivot table. You will see NaN where there was no data. This seems logical here, as you wouldn't expect temp_f values between -10 and 20 in the spring and summer.

```python
#bigfoot_valid.groupby(['classification','temp_bin','season'])['temp_f'].mean().unstack()
bigfoot_valid.pivot_table('temp_f', ['classification','temp_bin'], 'season')
```

The aggfunc keyword controls the type of aggregation, which uses the mean by default. You can use 'sum', 'mean', 'count', 'min', 'max', etc.
